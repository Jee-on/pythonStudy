```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
plt.style.use('classic')
x = np.linspace(0, 10, 100)

plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-0), color='g')
plt.plot(np.sin(x-1), color='g')
plt.plot(np.sin(x-2), color='g')
plt.plot(np.sin(x-3), color='g')
plt.plot(np.sin(x-4), color='g')
```

```python
""" 2. 선스타일 """

# plt.plot(np.sin(x-0), linestyle='solid')
# plt.plot(np.sin(x-1), linestyle='dashed')
# plt.plot(np.sin(x-2), linestyle='dotted')
# plt.plot(np.sin(x-3), linestyle='dashdot')

plt.plot(np.sin(x-0), linestyle='-')
plt.plot(np.sin(x-1), linestyle='--')
plt.plot(np.sin(x-2), linestyle=':')
plt.plot(np.sin(x-3), linestyle='-.')
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """
plt.plot(np.sin(x-0), 'mo-')
plt.plot(np.sin(x-1), 'cx--')
plt.plot(np.sin(x-2), 'yd-.')
plt.plot(np.sin(x-3), 'rx:')
```

```python
''' 4. 범례 '''

plt.plot(np.sin(x-0), 'mo-', label='first')
plt.plot(np.sin(x-1), 'cx--', label='second')
plt.plot(np.sin(x-2), 'yd-.', label='third')
plt.plot(np.sin(x-3), 'rx:', label='fourth')
plt.legend(loc=2)
```
