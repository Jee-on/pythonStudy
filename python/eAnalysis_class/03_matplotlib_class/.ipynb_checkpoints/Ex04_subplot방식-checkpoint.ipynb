```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(=?)   subplot    """
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) # == ax1 = plt.subplot(1,2,1) 
ax2 = fig.add_subplot(1,2,2)
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(100).cumsum())
```
