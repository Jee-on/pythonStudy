```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(그림틀=공간?)을 만들고 거기에 subplot을 추가 하는 방식 """
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1) # == ax1 = plt.subplot(1,2,1) 
ax2 = fig.add_subplot(1,2,2)
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(100).cumsum())
```

```python
""" 2. figure로 공간을 만들고 파이플롯(pyplot)의 subplot()으로 각각의 패널 지정 """
```

```python
""" 3. 간단하게 ax 객체를 이용하기 """
fig, ax = plt.subplots(1,2)
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```
