```python
import pandas as pd

df = pd.DataFrame( {'a':[10,20,30], 'b':[2, 4, 6]})
```

```python
# 1. 함수 정의
def my_fun1(x):
    return x**2

def my_fun2(x, n):
    return x**n

```

```python
# 2. apply 함수를 이용하여 모든 데이타에 각각 함수 적용하기
def my_fun1(x):
    return x**2
def my_fun2(x,n):
    return x**n

my_fun1(df['a'])
my_fun2(df['a'],2)

df.apply(my_fun1)
df['a'].apply(my_fun2(2))
```

```python
# [연습] 데이타 프레임의 각 열 단위로 데이타의 합을 출력
def my_fun3(col):
    sum=0
    for i in col:
        sum += i
    return sum

result = df.apply(my_fun3, axis='index')
result
```

```python
# 'a' 컬럼의 값이 10 이하는 'C', 20 이하는  'B', 30 이하는 'A'를 지정하고자 할 때
def func(score):
    grade = ''
    if( score <= 10) : grade='C'
    elif(score <= 20) : grade='B'
    elif(score <= 30) : grade='A'
    return grade
df['result'] = df['a'].apply(func)

df
```

## [연습]  고객의 결제정보에서 vip 선별

결재가 confirmed 된 상태이고 금액이 500원이상 구매한 고객들을 vip로 선별하고 그 외 고객들은 non-vip로 지정하여 grade 컬럼에 추가 저장한다.

```python
import pandas as pd

# index_col : 지정한 컬럼을 Row Index로 사용한다.

customer = pd.read_csv('./data/customer.csv', index_col = "Name")
customer
```

```python
def fun2(col):
    if col['price'] >= 500 and col['state']=='confirmed':
        grade = 'VIP'
    else:
        grade = 'non-VIP'
    return grade

customer['grade'] = customer.apply(fun2, axis=1 )
customer
```
